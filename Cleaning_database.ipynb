```python
#Revisando cuántos null hay
func_null = data.isnull().sum()
func_null
#El database es 796,315 entradas y de acuerdo con las columnas que tengo con null
#Decidí que me voy a descartar aquellas columnas con valores null de más de 150,000
#De igual manera, voy a remover los datos con valores null que se ven en las columnas folio,
#municipio_usuaria, colonia_usuaria y cp_usuaria
```

```python
#En esta lista voy a dejar los null con más de 150,000
cols_to_remove = list((func_null[func_null > 150000]).index)
cols_to_remove
```

```python
#Vamos a guardar un nuevo DF pero se van a quitar las columnas que ya seleccioné anteriormente
data_no_null = data.drop(cols_to_remove, axis = 1)
data_no_null.head()
```

```python
#Ahora voy a ver cuántos null quedan
data_no_null.isnull().sum()
```

```python
#En el caso de las columnas con null, voy a sustuir los valores con DESCONOCIDO
# Pero primero quiero ver el tipo de información que arrojan las columnas con valores null
#Revisando columna ESTADO_CIVIL
data_no_null.ESTADO_CIVIL.unique()
```

```python
#Ahora veo la columna escolaridad
data_no_null.ESCOLARIDAD.unique()
```

```python
data_no_null.ESTADO_USUARIA.unique()
```

```python
#Ahora veo la columna OCUPACION
data_no_null.OCUPACION.unique()
```

```python
#Ahora voy a sustutir los Nan con DESCONOCIDO de estas columnas
data_no_null[['ESTADO_CIVIL','OCUPACION','ESCOLARIDAD','TEMATICA_2','TEMATICA_3']] = data_no_null[['ESTADO_CIVIL','OCUPACION','ESCOLARIDAD','TEMATICA_2','TEMATICA_3']].replace(np.nan, "DESCONOCIDO")
```

```python
#Los registros con null que quedan, los voy a descartar
data_no_null.isnull().sum()
```

```python
#Ahora estoy quitando todos las filas donde hay null
data_no_null = data_no_null.dropna()
data_no_null.isnull().sum()
```

```python
### También faltó cambiar el tipo de CP_USUARIA a int
data_clean = data_no_null.astype({"CP_USUARIA":'int'})
data_clean.info()
#Ahora sí ya tengo mis datos como los necesito
```

```python
#Para todas las string voy a quitar la puntuación

data_clean['ESTADO_USUARIA'] = data_clean['ESTADO_USUARIA'].str.normalize('NFKD').str.encode('ascii',errors='ignore').str.decode('utf-8')
data_clean.head()
```

```python
##VOY A EXPLORAR LOS DATOS
#Con esta agrupación voy a sacar los registros por mes y año de cada estado,
data_clean.groupby(['AÑO_ALTA','MES_ALTA','ESTADO_USUARIA']).agg({'FOLIO':'count'}).reset_index()
```

```python
data_clean.groupby(['ESTADO_USUARIA','SEXO']).agg({'FOLIO':'count'}).reset_index()
```
